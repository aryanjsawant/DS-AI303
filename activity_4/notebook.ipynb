```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load data

data = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Convert TotalCharges to numeric

data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'].fillna(-1, inplace=True)

# Columns to encode

col_to_enc = ['gender','Partner','Dependents','PhoneService','MultipleLines',
'InternetService','OnlineSecurity','OnlineBackup','DeviceProtection',
'TechSupport','StreamingTV','StreamingMovies','Contract',
'PaperlessBilling','PaymentMethod','Churn']

le = LabelEncoder()

for col in col_to_enc:
    data[col + '_le'] = le.fit_transform(data[col])

# Define target

label = data['Churn_le']

# Drop original columns

data = data.drop(col_to_enc, axis=1)

# Train/test split

X_train, X_test, y_train, y_test = train_test_split(data, label, random_state=1)

# Logistic regression

lr = LogisticRegression(max_iter=1000, solver='liblinear')
lr.fit(X_train, y_train)
```
