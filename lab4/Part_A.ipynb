```python
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = sns.load_dataset('iris')
```

```python
df.columns
```

```python
df['sepal_length'].mean()
```

```python
setosa_sl = df[df['species'] == 'setosa']['sepal_length']
t_stat, p_value = stats.ttest_1samp(setosa_sl, 5.0)
t_stat, p_value
```

```python
versicolor_pw = df[df['species'] == 'versicolor']['petal_width']
virginica_pw = df[df['species'] == 'virginica']['petal_width']
t_stat, p_value = stats.ttest_ind(versicolor_pw, virginica_pw)
t_stat, p_value
```

```python
median_sl = df['sepal_length'].median()
df['sepal_length_cat'] = np.where(df['sepal_length'] < median_sl, 'Short', 'Long')

contingency = pd.crosstab(df['sepal_length_cat'], df['species'])
contingency
```

```python
chi2, p_value, dof, expected = stats.chi2_contingency(contingency)
chi2, p_value
```

```python
r, p_value = stats.pearsonr(df['sepal_length'], df['petal_length'])
r, p_value
```
